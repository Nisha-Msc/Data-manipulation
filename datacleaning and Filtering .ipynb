```python
unique=(['Million Dollar Baby - Single', 'Not Like Us',
       'I like the way you kiss me', ...,
       'Dil Meri Na Sune (From "Genius")', 'November Top 10 Songs',
       'Me Acostumbre (feat. Bad Bunny)'])
df['Album Name']=df['Album Name'].apply(lambda x:x if x in unique else np.nan)
df['Album Name'].unique()
df.dropna(subset=['Album Name'],inplace=True)
df['Album Name'].isnull().sum() 
```
